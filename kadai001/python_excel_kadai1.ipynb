```python
import openpyxl
from openpyxl.styles import Border, Side
import datetime

# 新しいワークブックを作成
workbook = openpyxl.Workbook()
sheet = workbook.active

# 指定した行に新しい行を挿入
sheet.insert_rows(1, 1)
sheet.insert_rows(2, 1)
sheet.insert_rows(7, 2)
sheet.insert_rows(13, 1)

# セルに、値「を設定
sheet['B2'].value = '請求書'

# セルに、値「を設定
sheet['B4'].value = '株式会社ABC'
sheet['F4'].value = 'No.'
sheet['G4'].value = '0001'
sheet['B5'].value = '〒101-0022 東京都千代田区神田練塀町300'
sheet['F5'].value = '日付'
sheet['B6'].value = 'TEL:03-1234-5678 FAX:03-1234-5678'
sheet['B7'].value = '担当者名:鈴木一郎 様'

# 指定した日時を取得する
date_time = datetime.datetime.strptime("2023-04-08", "%Y-%m-%d")
# G5 セルに代入
sheet["G5"] = date_time
sheet["G5"].number_format = "yyyy/mm/dd"  # 表示を 2024/04/08 に固定

# 明細
sheet["B10"] = "商品名"
sheet["C10"] = "数量"
sheet["D10"] = "単価"
sheet["E10"] = "金額"

# 商品A
sheet["B11"] = "商品A"
sheet["C11"] = 2
sheet["D11"] = 10000
sheet["E11"] = "=C11*D11"

# 商品B
sheet["B12"] = "商品B"
sheet["C12"] = 1
sheet["D12"] = 15000
sheet["E12"] = "=C12*D12"

# 合計
sheet["E13"] = "=SUM(E11:E12)"

# 小計（明細合計）
sheet["B15"] = "小計"
sheet["E15"] = "=SUM(E11:E12)"

# 消費税（10%）
sheet["B16"] = "消費税"
sheet["E16"] = "=E13*0.1"

# 合計
sheet["B17"] = "合計"
sheet["E17"] = "=E15+E16"

# 保存
workbook.save('kadai_001.xlsx')


```
